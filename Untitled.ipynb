```python
lines = []
with open('./input_data/a_an_example.in.txt') as f:
    lines = f.readlines()
 
data = []
for line in lines:
    data.append(line.split())
 
C = int(data[0][0])
P = int(data[0][1])
 
print("Contributors - ", C)
print("Projects - ", P)
index = 1
 
contribs = []
projects = []
 
for _ in range(C):
 
    name = data[index][0]
    skillCount = int(data[index][1])
 
    contrib = {'name': name, 'skills': []}
    index += 1
    for j in range(skillCount):
        skill = {'name': data[index][0], 'level': int(data[index][1])}
        index += 1
        contrib['skills'].append(skill)
 
    contribs.append(contrib)
 
for _ in range(P):
 
    proj = {
        'name': data[index][0],
        'days': int(data[index][1]),
        'score': int(data[index][2]),
        'bestBefore': int(data[index][3]),
        'roles': int(data[index][4]),
        'skills': [],
        'completed': False,
    }
    index += 1
    for i in range(proj['roles']):
        skill = {'name': data[index][0], 'level': int(data[index][1])}
        proj['skills'].append(skill)
        index += 1
 
    projects.append(proj)
 
 
# for p in projects:
#     print(p)
    
import pprint
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(projects)
print()
pp.pprint(contribs) 
```
